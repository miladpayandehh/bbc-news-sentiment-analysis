```python
from textblob import TextBlob
from newspaper import Article
```

```python
url = 'https://zharfapodcast.com/bbcnewssentiment.html'
article = Article(url)
article.download()
article.parse()
print(article.text)
```

```python
text = article.text
blob = TextBlob(text)
sentiment = blob.sentiment.polarity # -1 to +1
print(sentiment)
```
